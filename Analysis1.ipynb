### Handle Missing Data

```python
df.isnull().sum()/df.shape[0]*100
```

```python
#Changing data types for First and second term Gpa:-
df['First Term Gpa'] = df['First Term Gpa'].astype(float)
df['Second Term Gpa'] = df['Second Term Gpa'].astype(float)
```

Handling missing values for numerical column:-

```python
#Handling missing values for First Term Gpa and Second Term Gpa:-
df['First Term Gpa'].fillna(df['First Term Gpa'].mean(), inplace=True)
df['Second Term Gpa'].fillna(df['Second Term Gpa'].mean(), inplace=True)
```

```python
# Check missing values in the 'math_score' column
math_missing_percentage = df['Math Score'].isnull().mean() * 100
print(f"Missing values in math score column: {math_missing_percentage}%")
```

Handling missing values for categorical column:-

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df['Gender'].value_counts()
```

```python
#Removing the rows, where value ==3 :-
df = df[df['Gender'] != 3]
df['Gender'].value_counts()
```

```python
df.dtypes
```

```python
#Value counts for English Grade:-
df['English Grade'].value_counts()
```

```python

# Imputing missing values for English Grade with mode:-
df['English Grade'].fillna(df['English Grade'].mode()[0], inplace = True)
```

```python
# Imputing missing values for English Grade with mode:-
df['Prev Education'].fillna(df['Prev Education'].mode()[0], inplace = True)
```

```python
#Rechecking missing values for English Grade:-
df.isnull().sum()
```

```python
# Imputing missing values for First Language with mode:-
df['First Language'].fillna(df['First Language'].mode()[0], inplace = True)
```

```python
df.isnull().sum()
```

```python
#Handling missing values for Age Group:-
df['Age Group'].value_counts()
```

```python
df.dtypes
```

```python
print(df['Age Group'].isnull().sum())
df = df.dropna(subset=['Age Group'])
```

Detecting Outliers:-

Outliers for Prev Education:-

```python
# Detecting outliers in the 'prev education' column:-
invalid_values = df[~df['Prev Education'].isin([1, 2])]
print(f"Invalid values found:\n{invalid_values}")
```

As we can see  that our data ranges from 1 and 2 which represents 1-> Highschool and 2-> Postsecondary, and there are 88 rows which contain 0 value which represents them as an outliers.

Outliers for High School Average Mark.

Exploratory Data Analysis:-

```python
import plotly.express as px
```

```python
#Pie Chart for Gender Proportion:-
#Showing proportions for male and female:-
gender_mapping = {1: "Male", 2: "Female"}
df['gender'] = df['Gender'].map(gender_mapping)

fig = px.pie(
    df,
    names='gender',
    title="Gender Proportion"
)
fig.show()
```

```python
#creating a histogram the GPA distribution varies across different residency status:
import plotly.express as px

fig = px.histogram(
    df,
    x='First Term Gpa',
    color='Residency',
    barmode='overlay', 
    title="Distribution of First Term GPA by Residency",
    labels={'First Term Gpa': 'First Term Gpa'},
    opacity=0.6  
)
fig.show()
```

```python
#Here with this visualization we are trying to compare First Term Gpa and Second Term Gpa for each gender:-
fig = px.bar(df, 
             x='Gender', 
             y='First Term Gpa', 
             title="First Term GPA Distribution by Gender")
fig.show()

fig = px.bar(df, 
             x='Gender', 
             y='Second Term Gpa', 
             title="Second Term GPA Distribution by Gender")
fig.show()
```
